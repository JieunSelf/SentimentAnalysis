```python
import pandas as pd
PATH = '../corpus/4_tokens/tokens6_18_score.tsv'
df = pd.read_csv(PATH, encoding = 'utf-8', sep = '\t')
df[10:30]
```

```python
df['age'].value_counts().sort_index()
```

```python
# label 생성
# 6~11세 : 0, 12~18세 : 1
df['label'] = [0 if s <11 else 1 for s in df['age']]
df[10:30]
```

```python
df2 = df.replace({'senti_score':'None'},0)
df2[10:30]
```

```python
df2.dtypes
```

```python
df2['senti_score']  = pd.to_numeric(df2["senti_score"])
df2[10:30]
```

```python
df2.groupby(['label'], as_index=False).mean()
```

```python
df2.groupby(['label'], as_index=False).std()
```

```python
df2.groupby(['age'], as_index=False).mean() 
```

```python
# t-test
from scipy.stats import levene
young = df2['senti_score'][(df2.label == 0)]
old = df2['senti_score'][(df2.label == 1)]
levene(young, old) # Homogeneity of Variance Test
```

```python
# Welche's t-test
from scipy import stats
stats.ttest_ind(young, old, equal_var=False)
```
